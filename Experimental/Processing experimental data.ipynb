## Doing Non-linear square fitting to experimental data

```python
from glob import glob
from Lib_FS import log_scale, NLS_fit, linear_fit_Nob, log_tw
from Lib_rheology import chi_th, theta_loss, conv_U_F
from Lib_FDprocess import average_FDcurves, repulsiveFD
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
files = glob('*.ibw')  
tav, zav, dav, tipav, karr, invarr = average_FDcurves(files)

t, z, d, tip = repulsiveFD(tav, zav, dav)
```

## Performing the Non Linear Square Optimization

```python
t_res = 1.0e-4   #time resolution (inverse of sampling frequency)
t_exp = 1.0e-1  #total experimental time

R = 10.0e-9

tip_log, t_log_sim = log_scale(tip, t, t_res, t_exp) #Weighting time and tip arrays in logarithmic scale
F_log, _ = log_scale(d*karr[0], t, t_res, t_exp) #Weighting force array in logarithmic scale
Fdot = linear_fit_Nob(t_log_sim, F_log)    #Getting linear slope of force in time trace
chi_simul = 16.0*np.sqrt(R)/3*pow(tip_log,1.5)/Fdot     #according to eq 19, relation between chi and tip when force is assumed to be linear
#SIMULATION RESULTS#

#NON-LINEAR SQUARE FITTING
arms = 3
Jg_c, tau_c, J_c = NLS_fit(t, tip, d*karr[0], R, t_res, t_exp, arms, 1, 1.0e-9, 1.0e-9, 1.0e-3, 1.0e-6, 1.0e-2, 1.0e-5, 1.0e-1)
Jg_linear, tau_linear, J_linear = NLS_fit(t, tip, d*karr[0], R, t_res, t_exp, arms, 0, 1.0e-9, 1.0e-9, 1.0e-4, 1.0e-6, 1.0e-3, 1.0e-5, 1.0e-2)
Jg_a = np.zeros(arms)
Jg_a[:] = Jg_c
#NON-LINEAR SQUARE FITTING
```

### Doing plots to check visually the non-linear optimization

```python
plt.figure(1)
plt.plot(t_log_sim, 16.0/3.0*np.sqrt(R)*tip_log**1.5, 'k*',lw = 3.0,  label = 'Experimental Data')



data_fit = conv_U_F(t, d*karr[0], Jg_c, J_c, tau_c)
data_fit_log, time_fit_log = log_scale(data_fit, t, t_res, t_exp)
plt.plot(time_fit_log, data_fit_log, 'c', lw = 5.0, label = 'Non-linear squares fit, Non Linear')

linear_fit = conv_U_F(t, d*karr[0], Jg_linear, J_linear, tau_linear)
linear_fit_log, time_linear_log = log_scale(linear_fit, t, t_res, t_exp)
plt.plot(time_fit_log, linear_fit_log, 'r--', lw = 3.0, label = 'Non-linear squares fit, Linear')

plt.legend(loc=4)
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$time, \,s$', fontsize='20',fontweight='bold')
plt.ylabel(r'$\int_{0}^{t} U(t-\xi) F(\xi) d\xi, \,m^2$',fontsize='20',fontweight='bold')
plt.suptitle('Non-linear squares fit',fontsize='x-large',fontname='times new roman')

plt.figure(2)
omega = log_tw(2.0*np.pi*(1.0/t_exp), 2.0*np.pi*(1.0/t_res),20)
plt.plot(omega, theta_loss(omega, Jg_c, J_c, tau_c, 0.0), 'c', lw= 5, label = 'Fit Non Linear')
plt.plot(omega, theta_loss(omega, Jg_linear, J_linear, tau_linear, 0.0), 'r--', lw= 5, label = 'Fit Linear')
plt.legend(loc=4, fontsize=13)
plt.xlabel(r'$\omega, \,rad/s$', fontsize='20',fontweight='bold')
plt.ylabel(r'$\theta(\omega), \,deg$',fontsize='20',fontweight='bold')
plt.xscale('log')
plt.xlim(10,10**4)
```

```python
t_log = log_tw(t_res, t_exp)
omega = log_tw(1.0/t_exp/10.0, 1.0/t_res*10.0,20)

plt.figure(figsize=(15,5))
plt.subplot(121)
chi_5 = chi_th(t_log, Jg_linear, J_linear, tau_linear)
plt.plot(t_log_sim, chi_simul, 'r*', markersize=15, label=r'Experiment, see Eq.(14)')
plt.plot(t_log, chi_5, 'b', lw = 3.0, label=r'5-Voigt Fit, Eq. (9)')
plt.legend(loc=4, fontsize=13)
plt.xlabel(r'$time, \,s$', fontsize='20',fontweight='bold')
plt.ylabel(r'$\chi(t), \,Pa^{-1}s$',fontsize='20',fontweight='bold')
plt.xscale('log')
plt.yscale('log')

plt.subplot(122)
theta_5 = theta_loss(omega, Jg_linear, J_linear, tau_linear)
plt.plot(omega, theta_5, 'b', lw = 3.0, label=r'5-Voigt Fit')
plt.legend(loc=4, fontsize=13)
plt.xlabel(r'$\omega, \,rad/s$', fontsize='20',fontweight='bold')
plt.ylabel(r'$\theta(\omega),\,deg$',fontsize='20',fontweight='bold')
plt.xscale('log')
```
